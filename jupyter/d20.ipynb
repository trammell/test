```python
import pandas as pd
import matplotlib.pyplot as plt
import random


def d20():
    return random.randint(1,20)


df = pd.DataFrame({'d20': [d20() for i in range(1000000)]})
print(df.describe())
df.hist(bins=20, align="mid")    
```

```python
def d20a():
    return max(random.randint(1,20), random.randint(1,20))

df = pd.DataFrame({'d20a': [d20a() for i in range(10000000)]})
print(df.describe())
df.hist(bins=20)
```
